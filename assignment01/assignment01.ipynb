```python
def div7_mul5(num1 , num2):
    return [ num for num in range(num1, num2) if num % 7 == 0 and num % 5 !=0]
div7_mul5(2000,3200)
```

```python
fullname = input()

def reverse(s):
    s = s.split()
    l = 0
    r = len(s)-1
    while l < r:
        s[l] , s[r] = s[r] , s[l]
        l +=1
        r -=1
    return " ".join(s)
reverse(fullname)
```
